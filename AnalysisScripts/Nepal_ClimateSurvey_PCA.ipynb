# PCA Analysis


```python
#Importing Necessary Packages

import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
import seaborn as sns
from statsmodels.miscmodels.ordinal_model import OrderedModel
from pystout import pystout
from tabulate import tabulate
import latextable
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
#!pip install factor_analyzer
from factor_analyzer import FactorAnalyzer
```

```python
def confint(data):
    ci = 1.96 * np.std(data) / np.sqrt(len(data)-1)
    return ci
```

# Loading Data

```python
std_perceptions = pd.read_csv('ClimateRiskPerceptions.csv')
risk_perceptions = pd.read_csv('DisaggregatedRiskPerceptions.csv')
tempperceptions = pd.merge(risk_perceptions, std_perceptions, on='HHID')
cleanperceptions = tempperceptions.iloc[:,2:]

#HH 382 has NaN for Drought risk but high for everything else, so we assume drought will also be rated high
trainingdata = cleanperceptions.iloc[:,:-4]

trainingdata.iloc[382]['Droughts'] = 1.0
```

```python
#Categorizing Risk Perceptions
meanpercep = cleanperceptions['CompositeRisk'].mean()
cleanperceptions['Target'] = 'Low Risk'

cleanperceptions.loc[cleanperceptions['CompositeRisk'] > meanpercep, 'Target'] = 'High Risk'
cleanperceptions
```

```python
#Selecting only perceptions of future climate risk (not salience measures)
futurerisks = cleanperceptions.iloc[:,-12:]
futurerisks

trainingdata = futurerisks.iloc[:,0:6]
trainingdata.iloc[382]['Droughts'] = 1.0

trainingdata
```

```python
#Correlation Matrix for training data

table = trainingdata.corr()

print(tabulate(table, tablefmt='latex'))
print('\nTexttable Latex:')
```

# Factor Analysis

Documentation: https://www.datacamp.com/tutorial/introduction-factor-analysis

```python
#Normalize data
x = trainingdata.values
x = StandardScaler().fit_transform(x) # normalizing the features
```

```python
#Test suitability of data for factor analysis 
#Bartlett Sphericity tests if there is sufficient correlation between variables
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(trainingdata)
chi_square_value, p_value
```

```python
#KMO Test assesses proportion of variance among all variables - value > 0.6 is usually the threshold for proceeding
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(trainingdata)
kmo_model
```

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer()
fa.fit(trainingdata, 6)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev

#In general, should only keep number of factors equal to number of eigenvectors > 1
```

```python
# Create scree plot using matplotlib
plt.scatter(range(1,trainingdata.shape[1]+1),ev)
plt.plot(range(1,trainingdata.shape[1]+1),ev)
plt.axhline(1, ls='--')
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

```python
#Re-do factor analysis with number of factors kept
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer(n_factors=2,rotation='varimax', use_smc=True)
fa.fit(trainingdata)
print(pd.DataFrame(fa.loadings_,index=trainingdata.columns))
```

```python
#Variance explained by factors
print(pd.DataFrame(fa.get_factor_variance(),index=['Variance','Proportional Var','Cumulative Var']))
```

```python
#Re-scaling training data based on factor loadings
newvariables = fa.fit_transform(x)
newriskperceptions = pd.DataFrame(newvariables, columns=['Flood_Factor','Withering_Factor'])

newriskperceptions['HHID'] = std_perceptions['HHID']

newriskperceptions.to_csv('RiskPerceptions_2FactorAnalysis.csv')
```

# Normalize Data

```python
x = trainingdata.values
x = StandardScaler().fit_transform(x) # normalizing the features

```

```python
#Check data is normalized
np.mean(x), np.std(x)
```

```python
#Re-format normalized data into table of features
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalized_risk = pd.DataFrame(x,columns=feat_cols)

normalized_risk
```

# Principal Components Analysis

```python
#Standardizing Source Likert Ratings
pca_risk = PCA(n_components=2)
principalComponents_risk = pca_risk.fit_transform(x)
```

```python
#Summarizing data by principal components
principal_risk_Df = pd.DataFrame(data = principalComponents_risk
             , columns = ['principal component 1', 'principal component 2'])

principal_risk_Df
```

```python
#Explained Variance
print('Explained variation per principal component: {}'.format(pca_risk.explained_variance_ratio_))
```

```python
#Plotting Data along components

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Risk Dataset",fontsize=20)
targets = ['High Risk', 'Low Risk']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = cleanperceptions['Target'] == target
    plt.scatter(principal_risk_Df.loc[indicesToKeep, 'principal component 1']
               , principal_risk_Df.loc[indicesToKeep, 'principal component 2'], c = color, s=50)

plt.legend(targets,prop={'size': 15})
plt.show()
```
